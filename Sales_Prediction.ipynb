## OUTLIERS

```python
import seaborn as sns
import matplotlib.pyplot as plt
def outlier(col):
    print(data[col].describe())
    sns.set(rc={'figure.figsize':(25,10)})
    sns.boxplot(x=data[col])
    plt.show()
    a=data[col].quantile(0.25)
    b=data[col].quantile(0.75)
    print(f"Q1:{a},Q3:{b}")
    iqr=b-a
    print("IQR :",iqr)
    l=a-(1.5*iqr)
    u=b+(1.5*iqr)
    print(f"RANGE IS ({l},{u})")
    tenth_percentile = np.percentile(data[col], 10)
    ninetieth_percentile = np.percentile(data[col], 90)
    for i in data.index:
        if data.loc[i,col]<=l:
            data.loc[i,col]=tenth_percentile
        elif data.loc[i,col]>u:
            data.loc[i,col]=ninetieth_percentile
    sns.boxplot(x=data[col]).set(title="after")
    plt.show()
    print(data[col].describe())
```
